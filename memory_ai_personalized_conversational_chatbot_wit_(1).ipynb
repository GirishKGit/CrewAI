```python
# Properly format the conversation history by extracting 'candidates' field
def format_dialogue(dialogue_list):
    # Iterate through the list of dictionaries and extract 'candidates'
    conversation = []

    for dialogue in dialogue_list:
        if 'candidates' in dialogue:
            # Extract the candidates array and format it as a conversation
            candidates = dialogue['candidates']
            conversation.extend([f"User: {utterance.strip()}" for utterance in candidates])

    return "\n".join(conversation)

# Apply to the conversation history
dialogue_history = format_dialogue(first_row['utterances'])
print(dialogue_history)
```

```python
import re

# Manually clean and extract content from the raw 'utterances' string
utterances_raw = persona_chat_df.iloc[0]['utterances']

# Use a regex to extract the text between the quotes, focusing on actual utterances (ignoring metadata like 'array')
utterances_clean = re.findall(r"'(.*?)'", utterances_raw)

# Print the cleaned utterances to inspect
print("Cleaned utterances content:\n", utterances_clean)
```

```python
# Filter the cleaned utterances to remove non-conversational metadata
def filter_conversation(utterances_clean):
    filtered_conversation = []

    # Keywords to filter out
    metadata_keywords = ['candidates', 'history']

    # Go through each entry and keep only conversational content
    for utterance in utterances_clean:
        # Ignore metadata keywords
        if utterance.lower() not in metadata_keywords:
            filtered_conversation.append(utterance.strip())  # Add conversation, clean extra spaces

    return filtered_conversation

# Apply the filter to cleaned utterances
cleaned_conversation = filter_conversation(utterances_clean)

# Display the cleaned conversation
for entry in cleaned_conversation:
    print(entry)
```

```python
# Reformat the cleaned conversation for chatbot input
def format_for_chatbot(conversation):
    formatted_conversation = []
    for i, entry in enumerate(conversation):
        if i % 2 == 0:
            # Alternate between User and Bot (assuming alternating dialogue structure)
            formatted_conversation.append(f"User: {entry}")
        else:
            formatted_conversation.append(f"Bot: {entry}")

    return "\n".join(formatted_conversation)

# Format the conversation for chatbot
chatbot_ready_conversation = format_for_chatbot(cleaned_conversation)
print(chatbot_ready_conversation)
```

```python
# Filter out incomplete or garbled conversation entries
def filter_garbled_conversation(conversation):
    cleaned_conversation = []
    for entry in conversation:
        # Filter out entries that are too short or don't make sense (adjust threshold as needed)
        if len(entry) > 10 and '"' not in entry:
            cleaned_conversation.append(entry)
    return cleaned_conversation

# Apply the filter to cleaned conversation
cleaned_conversation = filter_garbled_conversation(cleaned_conversation)

# Format it for the chatbot again
chatbot_ready_conversation = format_for_chatbot(cleaned_conversation)
print(chatbot_ready_conversation)
```

```python
# Improved filter to catch incomplete or garbled entries
def filter_incomplete_conversation(conversation):
    cleaned_conversation = []

    for entry in conversation:
        # Filter out entries that are too short or incomplete
        if len(entry) > 10 and '"' not in entry:
            # Check if entry ends with a valid punctuation mark (., !, ?)
            if entry.strip()[-1] in ['.', '!', '?']:
                # Ensure there's a reasonable word count to avoid short garbled entries
                if len(entry.split()) > 2:
                    cleaned_conversation.append(entry.strip())  # Clean extra spaces and append

    return cleaned_conversation

# Apply the improved filter to the conversation
cleaned_conversation = filter_incomplete_conversation(cleaned_conversation)

# Format the cleaned conversation for chatbot
chatbot_ready_conversation = format_for_chatbot(cleaned_conversation)
print(chatbot_ready_conversation)
```

```python
import ast

# Create a dictionary for quick lookups of line text by lineID
lines_dict = pd.Series(lines.text.values, index=lines.lineID).to_dict()

# Function to match utterance IDs with their corresponding text using the dictionary
def get_conversation_text(utterance_ids, lines_dict):
    utterance_ids = ast.literal_eval(utterance_ids)  # Convert string representation of list to actual list
    conversation_text = [lines_dict[line_id] for line_id in utterance_ids if line_id in lines_dict]
    return conversation_text

# Apply the function to extract conversations in a vectorized way without warning
conversations.loc[:, 'conversation_text'] = conversations['utteranceIDs'].apply(lambda x: get_conversation_text(x, lines_dict))

# Preview the results
print(conversations[['characterID_1', 'characterID_2', 'movieID', 'conversation_text']].head())

```

```python
import json

# Define a function to reformat the movie dialogues correctly
def reformat_movie_dialogue(movie_dialogue_file):
    new_movie_dialogue = []

    with open(movie_dialogue_file, 'r') as f:
        for line in f:
            # Assuming the format is similar to {"prompt": "some text", "completion": "some text"}
            try:
                data = json.loads(line.strip())
                # Join characters if split into individual characters
                if len(data['prompt']) == 1 and len(data['completion']) == 1:
                    continue  # Skip character-level entries

                new_movie_dialogue.append(data)
            except json.JSONDecodeError:
                print(f"Skipping invalid line: {line.strip()}")
                continue

    # Write the reformatted data to a new file
    with open('reformatted_movie_dialogue.jsonl', 'w') as f_out:
        for entry in new_movie_dialogue:
            f_out.write(json.dumps(entry) + '\n')

    print(f"Successfully reformatted the movie dialogue data.")

# Run the function on the movie dialogue file
movie_dialogue_file = 'chatbot_fine_tuning_data.jsonl'
reformat_movie_dialogue(movie_dialogue_file)
```

```python
import gradio as gr
import cohere
import pickle
import os

# Step 1: Load the saved model, tokenizer, embeddings, FAISS index, and prompts from the pickle file
with open('chatbot_state_v1.pkl', 'rb') as f:
    data = pickle.load(f)

# Restore the saved components
tokenizer = data['tokenizer']
model = data['model']
embeddings = data['embeddings']
index = data['faiss_index']
prompts = data['prompts']

# Fetch the API key from the environment variable
api_key = os.getenv('COHERE_API_KEY')

# Set up the Cohere client using the API key
cohere_client = cohere.Client(api_key)

# Initialize an empty list to store the conversation history
conversation_history = []

# Function to retrieve the closest dialog based on a user's query
def retrieve_closest_dialog(query, tokenizer, model, index, prompts, device):
    inputs = tokenizer(query, return_tensors='pt', padding=True, truncation=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        cls_embedding = outputs.last_hidden_state[:, 0, :].cpu().numpy()

    # Perform FAISS search to find the closest match
    D, I = index.search(cls_embedding, k=1)

    # Retrieve the closest matching dialog from the dataset
    closest_dialog = prompts[I[0][0]]

    return closest_dialog

# Function to handle the chatbot conversation
def chatbot(user_input):
    global conversation_history

    # Step 1: Retrieve the closest dialog
    closest_dialog = retrieve_closest_dialog(user_input, tokenizer, model, index, prompts, device)

    # Step 2: Append user input and retrieved dialog to conversation history
    conversation_history.append(f"User: {user_input}")
    conversation_history.append(f"Previous conversation: {closest_dialog}")

    # Step 3: Build the prompt from the conversation history
    prompt = "\n".join(conversation_history[-6:])  # Limit to last 6 exchanges
    prompt += f"\nYour response:"

    # Step 4: Generate response using Cohere
    response = cohere_client.generate(
        model='command-r-plus-04-2024',
        prompt=prompt,
        max_tokens=100
    )

    # Step 5: Append the generated response to the conversation history
    generated_response = response.generations[0].text
    conversation_history.append(f"Bot: {generated_response}")

    # Return the generated response
    return generated_response

# Gradio Interface
def gradio_interface(user_input):
    return chatbot(user_input)

# Set up Gradio interface
interface = gr.Interface(fn=gradio_interface, inputs="text", outputs="text", title="Conversational Chatbot")

# Launch the Gradio app
interface.launch()
```
