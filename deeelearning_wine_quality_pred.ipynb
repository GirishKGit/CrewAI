```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize

# Load the dataset into a DataFrame (replace with your actual file path)
# df = pd.read_csv('path/to/winequality.csv')

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Display the first few rows of the dataset before normalization
print("\nData before normalization:")
print(df.head())

# Print column names to check if 'quality' exists
print("\nColumn names in the DataFrame:")
print(df.columns)

# Separate features and target
target_column = 'quality'  # Update to the correct target column name

if target_column in df.columns:
    X = df.drop([target_column, 'Id'], axis=1)  # Drop 'Id' if it's not a feature
    y = df[target_column]
else:
    raise KeyError(f"Column '{target_column}' not found in the DataFrame")

# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Convert the normalized data back to a DataFrame for easy viewing
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)

# Display the first few rows of the dataset after normalization
print("\nFeature values after normalization:")
print(X_normalized_df.head())

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Define models
models = {
    'Logistic Regression': LogisticRegression(multi_class='ovr'),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(probability=True)  # Set probability=True for ROC curve
}

# Train and evaluate each model
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")

    # Train the model
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)  # Probability estimates for each class

    # Evaluation
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 4))  # Adjust figure size for confusion matrix
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=df[target_column].unique(), yticklabels=df[target_column].unique())
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'{model_name} Confusion Matrix')
    plt.tight_layout()  # Adjust subplot parameters
    plt.show()

    # ROC Curve and AUC
    y_test_bin = label_binarize(y_test, classes=df[target_column].unique())
    plt.figure(figsize=(8, 4))  # Reduced figure size for ROC curve

    for i in range(y_test_bin.shape[1]):
        if len(set(y_test_bin[:, i])) > 1:  # Ensure both classes are present
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
            roc_auc = roc_auc_score(y_test_bin[:, i], y_prob[:, i])
            plt.plot(fpr, tpr, lw=2, label=f'{df[target_column].unique()[i]} ROC curve (area = %0.2f)' % roc_auc)
        else:
            print(f"Skipping ROC curve for class '{df[target_column].unique()[i]}' due to lack of variance in test set.")

    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} Receiver Operating Characteristic (ROC)')
    plt.legend(loc="lower right")
    plt.tight_layout()  # Adjust subplot parameters
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize

import altair as alt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Initialize Bokeh output in notebook
output_notebook()

# Load the dataset into a DataFrame (replace with your actual file path)
# df = pd.read_csv('path/to/winequality.csv')

# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Display the first few rows of the dataset before normalization
print("\nData before normalization:")
print(df.head())

# Print column names to check if 'quality' exists
print("\nColumn names in the DataFrame:")
print(df.columns)

# Separate features and target
target_column = 'quality'  # Update to the correct target column name

if target_column in df.columns:
    X = df.drop([target_column, 'Id'], axis=1)  # Drop 'Id' if it's not a feature
    y = df[target_column]
else:
    raise KeyError(f"Column '{target_column}' not found in the DataFrame")

# Normalize the features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Convert the normalized data back to a DataFrame for easy viewing
X_normalized_df = pd.DataFrame(X_normalized, columns=X.columns)

# Display the first few rows of the dataset after normalization
print("\nFeature values after normalization:")
print(X_normalized_df.head())

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Define models
models = {
    'Logistic Regression': LogisticRegression(multi_class='ovr'),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Support Vector Machine': SVC(probability=True)  # Set probability=True for ROC curve
}

# Train and evaluate each model
for model_name, model in models.items():
    print(f"\nTraining {model_name}...")

    # Train the model
    model.fit(X_train, y_train)

    # Predict on the test set
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)  # Probability estimates for each class

    # Evaluation
    print(f"\n{model_name} Classification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    # Matplotlib Confusion Matrix
    plt.figure(figsize=(8, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=sorted(df[target_column].unique()), yticklabels=sorted(df[target_column].unique()))
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'{model_name} Confusion Matrix (Matplotlib)')
    plt.tight_layout()
    plt.show()

    # Fix for Altair Confusion Matrix
    classes = sorted(df[target_column].unique())
    cm_df = pd.DataFrame(cm, index=classes[:cm.shape[0]], columns=classes[:cm.shape[1]])
    cm_melted = cm_df.reset_index().melt(id_vars='index', var_name='Predicted', value_name='Count')
    cm_melted.rename(columns={'index': 'True'}, inplace=True)
    alt_chart = alt.Chart(cm_melted).mark_rect().encode(
        x='Predicted:O',
        y='True:O',
        color='Count:Q'
    ).properties(
        title=f'{model_name} Confusion Matrix (Altair)',
        width=400,  # 8 inches * 50 pixels/inch = 400 pixels
        height=200  # 4 inches * 50 pixels/inch = 200 pixels
    )
    alt_chart.display()

    # ROC Curve and AUC
    y_test_bin = label_binarize(y_test, classes=df[target_column].unique())

    plt.figure(figsize=(8, 4))  # 8x4 dimension for ROC curve

    for i in range(y_test_bin.shape[1]):
        if len(set(y_test_bin[:, i])) > 1:  # Ensure both classes are present
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
            roc_auc = roc_auc_score(y_test_bin[:, i], y_prob[:, i])
            plt.plot(fpr, tpr, lw=2, label=f'{df[target_column].unique()[i]} ROC curve (area = %0.2f)' % roc_auc)
        else:
            print(f"Skipping ROC curve for class '{df[target_column].unique()[i]}' due to lack of variance in test set.")

    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'{model_name} Receiver Operating Characteristic (ROC) (Matplotlib)')
    plt.legend(loc="lower right")
    plt.tight_layout()
    plt.show()

    # Bokeh ROC Curve
    for i in range(y_test_bin.shape[1]):
        if len(set(y_test_bin[:, i])) > 1:  # Ensure both classes are present
            fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
            roc_data = pd.DataFrame({
                'fpr': fpr,
                'tpr': tpr,
                'class': df[target_column].unique()[i]
            })
            p = figure(title=f'{model_name} ROC Curve (Bokeh)', x_axis_label='False Positive Rate', y_axis_label='True Positive Rate', width=400, height=200)
            p.line(roc_data['fpr'], roc_data['tpr'], line_width=2, legend_label=f'{df[target_column].unique()[i]} ROC curve (area = %0.2f)' % roc_auc)
            p.line([0, 1], [0, 1], line_color='gray', line_dash='dashed')
            show(p)
```
